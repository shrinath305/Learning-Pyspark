# Spark dataframe are built on top of Spark SQL So we can filter using SQL 

```python
df.filter("Close < 100").show()
```

```python
df.filter("Close < 100").select (['Date','Open']).show()
```

```python
df.filter((df['Close'] < 200) & (df['Open'] >198)).show()
```

```python
df.filter(df['Low']==197.16).show()
```
